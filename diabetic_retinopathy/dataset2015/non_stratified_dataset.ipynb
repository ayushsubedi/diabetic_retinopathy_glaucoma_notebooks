```python
df = pd.read_csv('/content/gdrive/MyDrive/blurry_dark_ratios.csv')
```

```python
df.head()
```

```python
def new_path_fun(x):
  return "/content/"+("/".join(x.split('/')[3:]))
```

```python
new_path_fun(df.iloc[0].path)
```

```python
df['new_path'] = df.path.apply(new_path_fun)
```

```python
df
```

```python
clean_df = df[((df.dark>35) & (df.dark<150) & (df.blurry>25) & (df.blurry<250))]
```

```python
stratify_df = clean_df.copy()
```

```python
stratify_df = pd.concat([resample(stratify_df[stratify_df.level==0], replace=False, n_samples=1297, random_state=42),
          resample(stratify_df[stratify_df.level==1], replace=False, n_samples=1297, random_state=42),
          resample(stratify_df[stratify_df.level==2], replace=False, n_samples=1297, random_state=42),
          resample(stratify_df[stratify_df.level==3], replace=False, n_samples=1297, random_state=42),
          resample(stratify_df[stratify_df.level==4], replace=False, n_samples=1297, random_state=42)])
```

```python
stratify_df
```

```python
stratify_df.reset_index(drop=True, inplace=True)
```

```python
stratify_df
```

```python
SEED=42
train_x, valid_x, train_y, valid_y = train_test_split(stratify_df, stratify_df.level, test_size=0.20,
                                                      stratify=stratify_df.level, random_state=SEED)
```

```python
train_x
```

```python
def move_files_valid(df):
    location = "data_2015_stratify/val/"+str(df.level)+"/"+df.image+".jpg"
    shutil.copy(df.new_path, location)
```

```python
valid_x.apply(move_files_valid, axis=1)
```

```python
def move_files_train(df):
    location = "data_2015_stratify/train/"+str(df.level)+"/"+df.image+".jpg"
    shutil.copy(df.new_path, location)
```

```python
train_x.apply(move_files_train, axis=1)
```
